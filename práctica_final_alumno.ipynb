```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']
```

```python
# A. Una función que calcule el precio total de la compra

calculo_total = lambda a,b : reduce(lambda acum, producto: acum + b[producto], a, 0)

calculo_total(compra, articulos)
```

```python
# B. Una función que calcule el precio total de la compra con un descuento del 10%

descuento = lambda x,i: x - (x*i/100)

total_mas_descuento_10 = lambda a,b,c : descuento(calculo_total(a,b), c)

total_mas_descuento_10(compra, articulos, 10)
```

```python
# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€
descuento_mas_100 = lambda x: descuento_10(x) if x > 100  else x

total_mas_descuento_100 = lambda a,b : descuento_mas_100(calculo_total(a,b))

total_mas_descuento_100(compra, articulos)
```

```python
# D. Una función que calcule el precio total aplicando el IVA (21%)

iva = lambda x,i: x + (x*i/100)

total_con_iva = lambda a,b,c : iva(calculo_total(a,b), c)

total_con_iva(compra, articulos, 21)
```

```python
# E. Lista los artículos cuyo precio es superior a 20€

list(filter(lambda x: articulos[x] > 20, articulos))
```
